```python
## random date geenrator
import random
import time

def strTimeProp(start, end, format, prop):
    """Get a time at a proportion of a range of two formatted times.

    start and end should be strings specifying times formated in the
    given format (strftime-style), giving an interval [start, end].
    prop specifies how a proportion of the interval to be taken after
    start.  The returned time will be in the specified format.
    """

    stime = time.mktime(time.strptime(start, format))
    etime = time.mktime(time.strptime(end, format))

    ptime = stime + prop * (etime - stime)

    return time.strftime(format, time.localtime(ptime))


def randomDate(start, end, prop):
    return strTimeProp(start, end, '%m/%d/%Y %I:%M %p', prop)

randomDate("1/1/2008 1:30 PM", "1/1/2009 4:50 AM", random.random())
```

### Important:
when subtracting datetime.time objects, we can't simply say: t1-t2. Instead, we can convert them to datetime, by combining the same days date (for example today to both of them) then we can subtract them as datetime objects, then use total_seconds() to show the difference.

```python
# df['time_diff'] = df['contract_sent'] - df['signed']
df['contract_sent_2'] = df['contract_sent'].apply(lambda x: datetime.datetime.combine(datetime.date.today(), x))

df['signed_2'] = df['signed'].apply(lambda x: datetime.datetime.combine(datetime.date.today(), x))

df['diff_seconds'] = (df['signed_2'] - df['contract_sent_2']).apply(lambda x: x.total_seconds())  # returns seconds
df.drop(['signed_2','contract_sent_2'], axis=1)

df['diff_seconds'] = df['diff_seconds'].apply(lambda x : str(datetime.timedelta(seconds= x))) # converts seconds to hour/min/sec
```
